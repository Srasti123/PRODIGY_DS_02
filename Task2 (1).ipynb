```python
# Titanic Dataset Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset from seaborn
data = sns.load_dataset("titanic")

print("First 5 rows:\n", data.head())
print("\nDataset Info:")
print(data.info())
print("\nMissing Values:\n", data.isnull().sum())

# Basic Statistics
print("\nSummary Statistics:\n", data.describe(include="all"))

# Data Cleaning
# Fill missing 'age' with median
data['age'].fillna(data['age'].median(), inplace=True)

# Fill missing 'embarked' with most common value
data['embarked'].fillna(data['embarked'].mode()[0], inplace=True)

# Drop 'deck' due to many missing values
data.drop(columns=['deck'], inplace=True)

# Exploratory Data Analysis (EDA)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Age Distribution
sns.histplot(data['age'], kde=True, bins=30, ax=axes[0])
axes[0].set_title("Age Distribution")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Count")

# Survival by Gender
sns.barplot(x="sex", y="survived", data=data, ax=axes[1])
axes[1].set_title("Survival Rate by Gender")
axes[1].set_xlabel("Gender")
axes[1].set_ylabel("Survival Rate")

# Survival by Passenger Class
sns.barplot(x="pclass", y="survived", data=data, ax=axes[2])
axes[2].set_title("Survival Rate by Class")
axes[2].set_xlabel("Passenger Class")
axes[2].set_ylabel("Survival Rate")

plt.tight_layout()
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Survival by Embarkation Point
plt.figure(figsize=(6,4))
sns.barplot(x="embarked", y="survived", data=data)
plt.title("Survival Rate by Embarkation Point")
plt.xlabel("Embarked")
plt.ylabel("Survival Rate")
plt.show()

# Conclusion
print("\nâœ… Key Insights:")
print("- Younger passengers had higher survival chances.")
print("- Women survived at much higher rates than men.")
print("- 1st class passengers had the highest survival rate.")
print("- Embarkation point also influenced survival.")
```
